```python
#assigning the target vuariable
```

```python
y=d1["Salary"]
y
```

```python
x=d1[["YearsExperience"]]
x
```

```python
#spliting the train and test data
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y,test_size=0.30)
```

```python
le=LinearRegression()
le.fit(x_train,y_train)
```

```python
y_pred=le.predict(x_test)
y_pred
```

```python
mse=mean_squared_error(y_test, y_pred)
mse
```

```python
print("mean squared error:",mse.round(3))
print("ROOT MEAN SQUARE ERROR",np.sqrt(mse).round(3))
print("R2 score:",r2_score(y_test, y_pred).round(3))
```

```python
#visualize the linear line
plt.scatter(x_test, y_test,color="Red")
plt.plot(x_test,y_pred,color="blue")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Regression Line")
plt.show()
```

```python
# Apply square root transformation
x_sqrt_train, x_sqrt_test = np.sqrt(x_train), np.sqrt(x_test)
le.fit(x_sqrt_train, y_train)
y_pred_sqrt = le.predict(x_sqrt_test)
mse_sqrt = mean_squared_error(y_test, y_pred_sqrt)
```

```python
# Print and visualize square root transformation
print("\nModel with Square Root Transformation:")
print("Mean Squared Error:", mse_sqrt.round(3))
print("Root Mean Squared Error:", np.sqrt(mse_sqrt).round(3))
print("R2 Score:", r2_score(y_test, y_pred_sqrt).round(3))
```

```python
plt.scatter(x_sqrt_test, y_test, color="Red")
plt.plot(x_sqrt_test, y_pred_sqrt, color="blue")
plt.xlabel("Square Root of Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Line with Square Root Transformation")
plt.show()
```

```python
# Apply logarithmic transformation
x_log_train, x_log_test = np.log1p(x_train), np.log1p(x_test)
le.fit(x_log_train, y_train)
y_pred_log = le.predict(x_log_test)
mse_log = mean_squared_error(y_test, y_pred_log)
```

```python
# Print and visualize logarithmic transformation
print("\nModel with Logarithmic Transformation:")
print("Mean Squared Error:", mse_log.round(3))
print("Root Mean Squared Error:", np.sqrt(mse_log).round(3))
print("R2 Score:", r2_score(y_test, y_pred_log).round(3))
```

```python
plt.scatter(x_log_test, y_test, color="Red")
plt.plot(x_log_test, y_pred_log, color="blue")
plt.xlabel("Logarithm of Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Line with Logarithmic Transformation")
plt.show()
```

```python
# Apply exponential transformation
x_exp_train, x_exp_test = np.exp(x_train), np.exp(x_test)
le.fit(x_exp_train, y_train)
y_pred_exp = le.predict(x_exp_test)
mse_exp = mean_squared_error(y_test, y_pred_exp)
```

```python
# Print and visualize exponential transformation
print("\nModel with Exponential Transformation:")
print("Mean Squared Error:", mse_exp.round(3))
print("Root Mean Squared Error:", np.sqrt(mse_exp).round(3))
print("R2 Score:", r2_score(y_test, y_pred_exp).round(3))
```

```python
plt.scatter(x_exp_test, y_test, color="Red")
plt.plot(x_exp_test, y_pred_exp, color="blue")
plt.xlabel("Exponential of Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Line with Exponential Transformation")
plt.show()
```

```python
# Apply polynomial transformation
x_poly_train, x_poly_test = x_train**2, x_test**2
le.fit(x_poly_train, y_train)
y_pred_poly = le.predict(x_poly_test)
mse_poly = mean_squared_error(y_test, y_pred_poly)
```

```python
# Print and visualize polynomial transformation
print("\nModel with Polynomial Transformation:")
print("Mean Squared Error:", mse_poly.round(3))
print("Root Mean Squared Error:", np.sqrt(mse_poly).round(3))
print("R2 Score:", r2_score(y_test, y_pred_poly).round(3))
```

```python
plt.scatter(x_poly_test, y_test, color="Red")
plt.plot(x_poly_test, y_pred_poly, color="blue")
plt.xlabel("Square of Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Line with Polynomial Transformation")
plt.show()
```

```python
'''The simple linear regression model performs exceptionally
well with the lowest errors and the highest R2 score, 
suggesting that a linear relationship between years of experience 
and salary is a good fit for the data.
.'''
```

```python
'''Among the transformed models, the square root transformation and
logarithmic transformation provide good fits with lower errors 
compared to the exponential transformation.
'''
```

```python
'''The polynomial transformation introduces higher errors but still captures a
significant portion of the variance, indicating a relatively good fit'''
```
